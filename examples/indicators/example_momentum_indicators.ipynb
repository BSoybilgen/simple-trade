### Load Functions

```python
from simple_trade import compute_indicator, download_data
import pandas as pd
from simple_trade import IndicatorPlotter

# Set pandas display options for better output
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# Create a plotter instance
plotter = IndicatorPlotter()
```

### The Commodity Channel Index (CCI)

```python
# Step 1: Download data
symbol = 'MSFT'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
parameters = dict()
columns = dict()
parameters["window"] = 20
parameters["constant"] = 0.015
columns["high_col"] = 'High'
columns["low_col"] = 'Low'
columns["close_col"] = 'Close'
data = compute_indicator(
    data=data,
    indicator='cci',
    parameters=parameters,
    columns=columns
)

# Step 3: Plot the indicator
window = parameters["window"]
constant = parameters["constant"]
columns = [f'CCI_{window}_{constant}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=True, 
        title=f"{symbol} with CCI({window}, {constant})"
    )
```

### The Moving Average Convergence Divergence Index (MACD), Signal Line, and Histogram

```python
# Step 1: Download data
symbol = 'MSFT'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
parameters = dict()
columns = dict()
parameters["window_fast"] = 12
parameters["window_slow"] = 26
parameters["window_signal"] = 9
columns["close_col"] = 'Close'
data = compute_indicator(
    data=data,
    indicator='macd',
    parameters=parameters,
    columns=columns
)

# Step 3: Plot the indicator
window_fast = parameters["window_fast"]
window_slow = parameters["window_slow"]
window_signal = parameters["window_signal"]
columns = [f'MACD_{window_fast}_{window_slow}', f'Signal_{window_signal}', 
           f'Hist_{window_fast}_{window_slow}_{window_signal}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=True, 
        title=f"{symbol} with MACD({window_fast},{window_slow},{window_signal})"
    )
```
