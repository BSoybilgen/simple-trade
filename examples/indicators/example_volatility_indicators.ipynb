### Load Functions

```python
from simple_trade import compute_indicator, download_data
import pandas as pd
from simple_trade import IndicatorPlotter

# Set pandas display options for better output
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# Create a plotter instance
plotter = IndicatorPlotter()
```

### The Average True Range (ATR)

```python
# Step 1: Download data
symbol = 'GOOG'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
window=14
data = compute_indicator(
    data=data,
    indicator='atr',
    window=window,
)

# Step 3: Plot the indicator
columns = [f'ATR_{window}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=True, 
        title=f"{symbol} with ATR_{window}"
    )
```

### The Bollinger Bands (BOLLIN)

```python
# Step 1: Download data
symbol = 'GOOG'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
window=20
num_std=2
data = compute_indicator(
    data=data,
    indicator='bollin',
    window=window,
    num_std=num_std,
    close_col='Close',

)

# Step 3: Plot the indicator
columns = [f'BB_Middle_{window}', 
           f'BB_Upper_{window}_{num_std}', 
           f'BB_Lower_{window}_{num_std}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=False, 
        title=f"{symbol} with BOOLINGER ({window}, {num_std})"
    )
```

### The Keltner Channels (KELT)

```python
# Step 1: Download data
symbol = 'GOOG'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
ema_window=20
atr_window=10
atr_multiplier=2.0
data = compute_indicator(
    data=data,
    indicator='kelt',
    ema_window=ema_window, 
    atr_window=atr_window, 
    atr_multiplier=atr_multiplier,
    high_col='High',
    low_col='Low',
    close_col='Close',
)

# Step 3: Plot the indicator
columns = [f'KELT_Middle_{ema_window}_{atr_window}_{atr_multiplier}', 
           f'KELT_Upper_{ema_window}_{atr_window}_{atr_multiplier}', 
           f'KELT_Lower_{ema_window}_{atr_window}_{atr_multiplier}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=False, 
        title=f"{symbol} with KELT ({ema_window}, {atr_window}, {atr_multiplier})"
    )
```
