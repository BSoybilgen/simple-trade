### Load Functions

```python
from simple_trade import compute_indicator, download_data
import pandas as pd
from simple_trade import IndicatorPlotter

# Set pandas display options for better output
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# Create a plotter instance
plotter = IndicatorPlotter()
```

### The Accumulation/Distribution Line (ADLINE)

```python
# Step 1: Download data
symbol = 'GOOG'
start = '2024-01-01'
end = '2024-12-31'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
data = compute_indicator(
    data=data,
    indicator='adline',
    high_col='High',
    low_col='Low',
    close_col='Close',
    volume_col='Volume'
)

# Step 3: Plot the indicator
columns = ['ADLINE']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=True, 
        title=f"{symbol} with ADLINE"
    )
```

### The Chaikin Money Flow (CMF)

```python
symbol = 'GOOG'
start = '2024-01-01'
end = '2025-01-01'
interval = '1d'
print(f"\nDownloading data for {symbol}...")
data = download_data(symbol, start, end, interval=interval)

# Step 2: Calculate indicator
period = 20
data = compute_indicator(
    data=data,
    indicator='cmf',
    period=period,
    high_col='High',
    low_col='Low',
    close_col='Close',
    volume_col='Volume'
)

# Step 3: Plot the indicator
columns = [f'CMF_{period}']
fig = plotter.plot_results(
        data,
        price_col='Close',
        column_names=columns,
        plot_on_subplot=True, 
        title=f"{symbol} with CMF_{period}"
    )
```
